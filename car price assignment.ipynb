Let's make a dataframe for applying model, consisting only columns which have significant impact on Price.

```python
cars_lr = cars_lr = cars[[ 'fueltype','carbody', 'drivewheel','wheelbase',
                  'curbweight', 'enginetype', 'cylindernumber', 'enginesize', 'boreratio','horsepower', 'carlength','carwidth','price']]
```

```python
cars_lr.head()
```

### Now, we need to convert categorical data into dummy variables. We can use onehotencoder or pd.getdummies
### Here we create a column with dummy numercial values for the categorical column fueltype. Then we concat this column with original cars_lr and drop the categorical column from the table. We will repeat the same step (create dummy,concat with table, remove the column whose dummy is created) to all categorical columns

```python
temp=pd.get_dummies(cars_lr['fueltype'],drop_first=True) #drop_first true : if gas then 1 else 0, where 0 signifies diesel
                                                         #therefore we won't need additional column for diesel 
temp
```

```python
cars_lr=pd.concat([cars_lr,temp],axis=1)  #axis=1 means column wise join
cars_lr.drop(['fueltype'],axis=1,inplace=True)

```

```python
cars_lr.head()
```

### now we do the same with drivewheel column

```python
cars_lr=pd.concat([cars_lr,pd.get_dummies(cars_lr['drivewheel'],drop_first=True)],axis=1)
cars_lr.drop(['drivewheel'],axis=1,inplace=True)
```

```python
cars_lr # we can see that drivewheel categorical column has been removed, and the dummy variables have been created
```

#### for enginetype

```python
cars_lr=pd.concat([cars_lr,pd.get_dummies(cars_lr['enginetype'],drop_first=True)],axis=1)
cars_lr.drop(['enginetype'],axis=1,inplace=True)
```

```python
cars_lr.head()
```

#### for carbody

```python
cars_lr=pd.concat([cars_lr,pd.get_dummies(cars_lr['carbody'],drop_first=True)],axis=1)
cars_lr.drop(['carbody'],axis=1,inplace=True)
```

```python
cars_lr.head()
```

#### for cylindernumber

```python
cars_lr=pd.concat([cars_lr,pd.get_dummies(cars_lr['cylindernumber'],drop_first=True)],axis=1)
cars_lr.drop(['cylindernumber'],axis=1,inplace=True)
```

```python
cars_lr.head()
```

#### For easier understanding, let's shift Price column to end of table

```python
p=cars_lr['price']
p
```

```python
cars_lr.drop(['price'],axis=1,inplace=True)
```

```python
cars_lr=pd.concat([cars_lr,p],axis=1)
cars_lr.head()
```

Our table is now ready.
Let's make X and y for training and testing

```python
cars_lr.shape
```

```python
X=cars_lr.iloc[:,:-1]# all rows,all columns upto last (not included)
X.head()
```

```python
y=cars_lr.iloc[:,26:27] # all rows , only select last column [:,-1] gives us 1d array,while 26:27 gives us 2d column,which can be used for visualization
y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=50)
```

```python
X.shape
```

```python
X_train.shape
```

```python
y.shape
```

```python
y_train.shape
```

```python
from sklearn.linear_model import LinearRegression

lr=LinearRegression()
lr.fit(X_train,y_train)
```

```python
y_test.head()
```

```python
lr.predict(X_test)
```

```python
lr.score(X_test,y_test)
```

We can see that our model has an accuracy of 83%
